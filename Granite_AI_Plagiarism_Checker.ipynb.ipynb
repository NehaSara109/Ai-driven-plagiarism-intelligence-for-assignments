```python
from ibm_watsonx_ai import APIClient
from ibm_watsonx_ai.foundation_models import ModelInference

# Step 1: Setup Watsonx client
creds = {
    "apikey": "pk2YEToNPRLcS8uYUxX2oCjLuxqE7pWMzHotRRtxiiqU",  # Replace with your API key
    "url": "https://us-south.ml.cloud.ibm.com"
}
project_id = "3e4a5e2c-d721-4e82-9831-ffa008aade73"  # Replace with your Watsonx project ID

client = APIClient(credentials=creds)

model = ModelInference(
    model_id="ibm/granite-13b-instruct-v2",
    project_id=project_id,
    credentials=creds
)

# Step 2: Historical submissions (simulated)
historical_submissions = [
    "Recursion allows a function to solve smaller parts of a problem until a base case is met.",
    "The recursive approach simplifies problems by dividing them into subproblems.",
    "A recursive function keeps calling itself until a stopping condition is reached."
]

# Step 3: New student submission
new_submission = "Recursive functions break down tasks and repeatedly call themselves until a base case is satisfied."

# Step 4: Compare new submission with historical ones
match_found = False
print("Comparing with historical submissions...\n")
for i, past in enumerate(historical_submissions):
    comparison_prompt = f"""Are the following two texts paraphrased versions of the same idea?

    Text 1: {past}
    Text 2: {new_submission}

    Reply with Yes or No and explain briefly."""
    
    response = model.generate(prompt=comparison_prompt)
    generated_text = response['results'][0]['generated_text']
    
    print(f"Comparison with Past Submission {i+1}:\n{generated_text}\n")
    
    if "yes" in generated_text.lower():
        match_found = True

# Step 5: Simulate instructor feedback
feedback_prompt = f"""
As an instructor, you're reviewing the following assignment.
Does it look suspicious compared to what students usually submit?
Explain your reasoning briefly.

Assignment: "{new_submission}"
"""

instructor_response = model.generate(prompt=feedback_prompt)
feedback_text = instructor_response['results'][0]['generated_text']

print("Instructor Feedback:\n", feedback_text)

# Step 6: Final flagging logic
print("\nFinal Verdict:")
if match_found or "yes" in feedback_text.lower():
    print("The assignment is flagged as suspicious/plagiarized.")
else:
    print("The assignment appears to be original.")




   
```
